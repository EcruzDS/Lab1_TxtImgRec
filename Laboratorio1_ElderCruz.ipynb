##### **1.3.1 Elegir valor: 1, 2, 3, 10, 20 ,30**

```python
color = 1
```

```python
def ajustar_colores(img, color):
    # Crear una copia de la imagen para no modificar la original
    img_salida = np.zeros_like(img)
    
    alto = img.shape[0]
    ancho = img.shape[1]

    for i in range(0, alto):
        for j in range(0, ancho):
            pixel = img[i, j]

            blue = pixel[2]
            green = pixel[1]
            red = pixel[0]

            if color == 1:
                img_salida[i, j] = [blue, 0, 0]
            elif color == 2:
                img_salida[i, j] = [0, green, 0]
            elif color == 3:
                img_salida[i, j] = [0, 0, red]
            elif color == 10:
                img_salida[i, j] = [0, green, red]
            elif color == 20:
                img_salida[i, j] = [blue, green, 0]
            elif color == 30:
                img_salida[i, j] = [blue, 0, red]

    return img_salida
```

#### **1.4 Resultado de monocromo**

```python
img_salida = ajustar_colores(img, color)

cv2.imwrite("monocromo.jpg",img_salida)
monocromo = cv2.imread("monocromo.jpg")
monocromo = cv2.cvtColor(monocromo, cv2.COLOR_BGR2RGB) #Conversion de imagen a RBG
plt.imshow(monocromo)
plt.show()
```

### **Problema 4**
#### Cree una función que dada una imagen, muestre el histograma de cada canal de color y el de escala de grises (utilice un *promedio aritmético* para su escala de grises, no puede usar funciones de opencv), sus histogramas deben incluir una línea vertical la cual muestre el valor de la media de la distribución.

```python

#Separar canales de color
red_channel = img3[:, :, 0]
green_channel = img3[:, :, 1]
blue_channel = img3[:, :, 2]

#Conversión de imagen a escala de grises usando promedio aritmético
gray_image = (red_channel.astype(float) + green_channel.astype(float)+ blue_channel.astype(float))/3
gray_image = gray_image.astype(np.uint8)

#Calcular y mostrar los histogramas de cada canal de color y el de escala de grises, incluyendo la media
channels = {'Red': red_channel, 'Green': green_channel, 'Blue': blue_channel, 'Gray': gray_image}

plt.figure(figsize=(10, 8))


for i, (name, channel) in enumerate(channels.items(), 1):
    # Calcular el histograma
    histogram, bin_edges = np.histogram(channel, bins=256, range=(0, 255))
    
    # Calcular la media
    mean_value = np.mean(channel)
    
    # Graficar el histograma
    plt.subplot(2, 2, i)
    plt.plot(bin_edges[0:-1], histogram, color=name.lower())
    plt.axvline(mean_value, color='k', linestyle='dashed', linewidth=1)
    plt.title(f'{name} Histograma de canales')
    plt.xlabel('Valor de pixeles')
    plt.ylabel('Frecuenca')
    plt.legend([f'Promedio: {mean_value:.2f}'])

plt.tight_layout()
plt.show()
```

### **Problema 5**
#### Investigue en qué consiste el enfoque de la escala de grises ponderado, luego de esto implemente una función que dada una imagen, realice una escala de grises ponderada (notar que no existe una solución única).

#### **Enfoque de escala de grises ponderado**
#### Consiste en convertir una imagen a color en escala de grises usando una combinación ponderada de valores de los canales rojo, verde, azul.  A diferencia del método de *promedio aritmético simple* donde cada canal contribuye por igual al valor en escala de grises, en este método, se asignan diferentes pesos a cada canal para reflejar cómo percibe el ojo humano la intensidad de cada color.

#### El *método ponderado* más comunmente usado en la conversión a escala de grises es el utilizado en la fórmula ITU-R BT.601, que asigna los siguientes pesos. Rojo: 0.299, verde: 0.587 y Azul: 0.114

```python
#Imagen Original
plt.imshow(img)
plt.show()
```

```python
#Separar canales de color

red_channel = img[:, :, 0]
green_channel = img[:, :, 1]
blue_channel = img[:, :, 2]

#aplicar conversión a escala de grises ponderada

gray_image = 0.299 * red_channel + 0.587 * green_channel + 0.114 * blue_channel
gray_image = gray_image.astype(np.uint8)

img_salida = gray_image

cv2.imwrite("ponderada.jpg",img_salida)
imgponderada = cv2.imread("ponderada.jpg")
imgponderada = cv2.cvtColor(imgponderada, cv2.COLOR_BGR2RGB) #Conversion de imagen a RBG
plt.imshow(imgponderada)
plt.show()
```

#### **Espacio de color HSV**
#### *Hue, Saturation, Value (HSV)* consiste en un modelo que se usa para describir y representar colores de una manera que es más intuitiva para los seres humanos que el espacio color RGB *(Rojo, verde, azul)*


#### *Hue (Matiz):* Representa el tipo de color. Se mide en grados en un círculo de color y va de 0° a 360°. Cada valor de matiz corresponde a un color específico:

#### - 0° corresponde al rojo.
#### - 120° corresponde al verde.
#### - 240° corresponde al azul.

#### *Saturation (Saturación):* Representa la intensidad o pureza del color. Va de 0% a 100%. Un color con 100% de saturación es puro (sin mezcla de blanco), mientras que un color con 0% de saturación es un tono de gris.

#### *Value (Valor o Brillo):* Representa el brillo del color. Va de 0% a 100%. Un color con 0% de valor es negro, y un color con 100% de valor es el más brillante posible para el matiz y saturación dados.

#### **Mapeo de Colores al Espacio HSV**
#### Para convertir un color del espacio RGB al espacio HSV, se utilizan las siguientes fórmulas:

#### 1. Normalizar los valores RGB dividiéndolos por 255 (para que estén en el rango de 0 a 1).

#### 2. *Calcular el valor máximo (𝐶𝑚𝑎𝑥), el valor mínimo (𝐶𝑚𝑖𝑛) y la diferencia (Δ=𝐶𝑚𝑎𝑥−𝐶𝑚𝑖𝑛).*

#### 3. *Calcular el Hue (H):*
#### Si Δ=0, entonces 𝐻=0

#### Si 𝐶𝑚𝑎𝑥=𝑅′, entonces 𝐻=60×((𝐺′−𝐵)/Δmod6)
#### Si 𝐶𝑚𝑎𝑥=𝐺′, entonces 𝐻=60×((𝐵′−𝑅′)/Δ+2)
#### Si 𝐶𝑚𝑎𝑥=𝐵′, entonces 𝐻=60×((𝑅′−𝐺′)/Δ+4)

#### *Calcular la Saturation (S):*
#### Si 𝐶𝑚𝑎𝑥=0, entonces 𝑆=0
#### Si 𝐶𝑚𝑎𝑥≠0, entonces 𝑆=Δ/𝐶𝑚𝑎𝑥

#### *Calcular el Value (V)*
#### 𝑉 = 𝐶𝑚𝑎𝑥

#### El resultado 𝐻 estará en el rango de 0° a 360°, mientras que 𝑆 y 𝑉 estarán en el rango de 0 a 1.
